```python
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[1,2],[3,4]])

print("Array 1:",arr1)
print("Array 2:",arr2)

print("Array 1 > Array 2", np.greater(arr1,arr2))
print("Array 1 >= Array 2", np.greater_equal(arr1,arr2))
print("Array 1 < Array 2", np.less(arr1,arr2))
print("Array 1 <= Array 2", np.less_equal(arr1,arr2))
print("Array 1 == Array 2", np.equal(arr1,arr2))
print("Array 1 != Array 2", np.not_equal(arr1,arr2))
```

```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1,1,1,2,3,4,2,4,3,3])

print("X array:",x)
print("Y array:",y)

x1 = np.bincount(x)
max_x1 = max(x1)
print("x1 array:",x1)

y1 = np.bincount(y)
max_y1 = max(y1)
print("y1 array:",y1)

print("Max freq. element of array x are:")
for i in range(len(x1)):
  if x1[i]==max_x1:
    print(i, end=" ")

print("\nMax freq. element of array y are:")
for i in range(len(y1)):
  if y1[i]==max_y1:
    print(i, end=" ")
```

```python
n_array = np.array([[55,25,15],[30,44,2],[11,45,77]])
print("The array is:",n_array)

dsum=0
for i in range(len(n_array)):
  for j in range(len(n_array[0])):
    if i == j:
      dsum = dsum + n_array[i][j]

print("\nThe sum of diagonal elements is: ",dsum)

eval, evect = np.linalg.eig(n_array)
print("\nThe eigen values of the matrix are:", eval)
print("\nThe eigen vectors of the matrix are:", evect)

invArr = np.linalg.inv(n_array)
print("\nThe inverse of the array is:",invArr)

detArr = np.linalg.det(n_array)
print("\nThe determinant of the array is:",detArr)
```
