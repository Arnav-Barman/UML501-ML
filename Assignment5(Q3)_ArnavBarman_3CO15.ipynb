```python
from sklearn.datasets import load_boston
boston=load_boston()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn import metrics
```

```python
boston
```

```python
bstn = pd.DataFrame(boston.data)
bstn.head()
```

```python
bstn.columns=boston.feature_names
```

```python
bstn.head()
```

```python
boston.data.shape
```

```python
boston.target.shape
```

```python
bstn["Price"]=boston.target
```

```python
bstn.head()
```

```python
X=bstn.iloc[:,:-1] ## independent features
y=bstn.iloc[:,-1] ## dependent features
```

```python
x_train, x_test, y_train, y_test = train_test_split(boston.data, boston.target)
```

```python
#LassoCV
lcv = linear_model.LassoCV(alphas = np.logspace(-3, -1, 30), normalize = True)
lcv.fit(x_train,y_train)
#RidgeCV
rcv = linear_model.RidgeCV(alphas = np.logspace(-3, -1, 30), normalize = True)
rcv.fit(x_train, y_train)
```

```python
ypred_lcv_train = lcv.predict(x_train)
ypred_lcv_test = lcv.predict(x_test)

ypred_rcv_train = rcv.predict(x_train)
ypred_rcv_test = rcv.predict(x_test)
```

```python
print("RidgeCV MSE (test): ", metrics.mean_squared_error(y_test,ypred_rcv_test))
print("RidgeCV MSE (train): ", metrics.mean_squared_error(y_train,ypred_rcv_train))
print("LassoCV MSE (test): ", metrics.mean_squared_error(y_test,ypred_lcv_test))
print("LassoCV MSE (train): ", metrics.mean_squared_error(y_train,ypred_lcv_train))
```

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits=10,random_state=None)
```

```python
msel = []
mser = []

for train_index , test_index in kfold.split(X):
    X_train_1 , X_test_1 = X.iloc[train_index,:],X.iloc[test_index,:]
    y_train_1 , y_test_1 = y[train_index] , y[test_index]
     
    lcv.fit(X_train_1,y_train_1)
    pred_values_lcv = lcv.predict(X_test_1)
    
    rcv.fit(X_train_1,y_train_1)
    pred_values_rcv = rcv.predict(X_test_1)
     
    mse_lcv = metrics.mean_squared_error(y_test_1,pred_values_lcv)
    mse_rcv = metrics.mean_squared_error(y_test_1,pred_values_rcv)
    msel.append(mse_lcv)
    mser.append(mse_rcv)
     
avg_msel_score = sum(msel)/10
avg_mser_score = sum(mser)/10
```

```python
print('MSE of each fold LCV - {}'.format(msel))
print('Avg MSE LCV : {}'.format(avg_msel_score))

print('MSE of each fold RCV - {}'.format(mser))
print('Avg MSE RCV : {}'.format(avg_mser_score))
```
